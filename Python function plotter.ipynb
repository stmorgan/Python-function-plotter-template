# Python Function Plotter Template

```python
import math
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def f(x):
    return np.log(x) + x**3  + np.exp(x) # Modify the function here! 

x_start = 1 # Start x value
x_end = 10 # End x value

#-------------------------------------------------------------------
x = np.linspace(x_start, x_end, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function Plotter')
```
